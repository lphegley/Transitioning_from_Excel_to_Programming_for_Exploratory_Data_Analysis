<h2> What is Exploratory Data Analysis (EDA)? <h2>

Exploratory Data Analysis is all about digging into the data and looking around in it instead of jumping right into the work of data analysis with no forethought. Emphasis is placed on looking at the data through different visual models to come to an understanding of its contents. The intent is to look with an open mind at what the data is telling you, not to look for confirmation of your pre-existing hypothesises. Visaulizations should make us accept what the data contains, even if we expect or hope it to appear a certain way. Sticking to only planned analysis takes away the ability to be suprised by the contents in data. EDA is not the same as data analysis, but rather the first step in beginning the process of it.  

Basic ways of looking at data include univariate distributions of data and bivariate relationships between variables. There are much more complex ways of doing EDA, which can be found by reading Hartwig and Dearing's (1979) *Exploratory Data Analysis* or John Tukey's (1977) *Exploratory Data Analysis*.

```python
# creates a list of lists
with open ('ucr_index.csv', 'r') as infile: 
    csv_in = csv.reader(infile)
    data = [] #creates an empty list
    for row in csv_in:
        data.append(row) #each row in the data is appended to the list
print(data)
```

<h4> 3.1 Univariate Analysis and Bivariate Analysis <h4>

Univariate analysis is when one variable at a time is analyized. This analysis is usually summerized in the form of a count. This is often visualized using frequency tables, bar graphs, and histograms. Bivariate anaylsis looks at the relationship between two variables. 

Types of variables:  
catagorical variables - data that can be divided into definite groups. Example: Married (Yes or No)   
ordinal variables - data that can be divided into groups that contain order. Example: Grades (A, B, C, D, F)  
numerical or continuous variables - data that can take up any value within a given range. Example: Population

```python
#making individual arrays for certain columns so they can be represented in graphs
year = []
unique_year = []

for individual_list in array_data:
        data_point = individual_list[0]
        year.append(data_point)
        
unique_number = set(year)
for number in unique_number:
    if number == 'year':
        pass
    else:
        unique_year.append(number)
        
unique_year = sorted(unique_year) #sorts numerically or alphabethically
print(unique_year)
```

```python
#making individual arrays for certain columns so they can be represented in graphs
county_name = []
unique_county = []

for individual_list in array_data:
        data_point = individual_list[3] #its in the third place of the list 
        county_name.append(data_point)
        
set_county_name = set(county_name)

for county in set_county_name:
    if county == 'county_name':
        pass
    else:
        unique_county.append(county)
        
unique_county = sorted(unique_county)
print(unique_county)
```

```python
#for each county, give us the average murder rate from 1982-2018
murder_county = df.groupby('county_name')['murder'].mean()
print(murder_county)

#transform from a panda's df into a numpy array
murder_county = murder_county.to_numpy()

#make the array an integer
murder_county = murder_county.astype('int64')

print(murder_county)

county_name = sorted(set(df['county_name']))
print(county_name)
```

```python
# for each county, give us the average aggrevated assault rate between 1982-2018
pop_county = df.groupby('county_name')['population'].mean()
print(pop_county)

pop_county = pop_county.to_numpy()

#make the array an integer
pop_county = pop_county.astype('int64')
print(pop_county)
```

```python
# to normalize the information, murders are going to be per capita
#outcome is number of murders per 100,000 people

for i in county_name: 
        murder_per_capita = murder_county/pop_county * 100000
        
murder_per_capita = np.round(murder_per_capita, 4)

print(murder_per_capita)
```

```python
#with the information we have, we can represent the average per capita murders per Illinois county in a horizontal bar chart
x = murder_per_capita
y = county_name
plt.barh(y,x, align = 'center')
plt.ylabel('County name', fontsize = 16)
plt.xlabel('Murder per 100,000 residents', fontsize = 16)
plt.title('Average Murder (per capita) per Illinois County from 1982-2018', fontsize = 20)
plt.rcParams['figure.figsize'] = (20,20)
plt.figure(dpi = 300)
    
plt.show()
```

```python
#we can also look at how arson counts for Champaign county changed during the time the data was collected
arson_count_Champaign = ucr_county_name.loc['Champaign', 'arson']

x = unique_year
y = arson_count_Champaign

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'r')
axes.set_xlabel('Years', fontsize = 16)
axes.set_ylabel('Arson Count for Champaign County', fontsize = 16)
axes.set_title('Annual Arson Count for Champaign County', fontsize = 20)

plt.show()
```

```python
# for each county, give us the average aggrevated assault rate between 1982-2018
aa_rate_county = df.groupby('county_name')['aggravated_assault_rate'].mean()
print(aa_rate_county)

aa_rate_county = aa_rate_county.to_numpy()

#make the array an integer
aa_rate_county = aa_rate_county.astype('int64')
print(aa_rate_county)
```

```python
#we can find out the aggrevated assault rate per 1,000 people
for i in county_name: 
        aa_per_capita = aa_rate_county/pop_county * 1000
        

aa_per_capita = np.round(aa_per_capita, 2)

print(aa_per_capita)
```

```python
fig1, ax = plt.subplots()

x = county_name
y = aa_per_capita
plt.barh(x, y) 
plt.title('Aggrevated assault average for each county 1982-2018')
plt.xlabel('Aggrevated Assault Rate per 1,000 residents')
plt.ylabel('County')
plt.figure(figsize = (5,10),dpi = 300)
ax.set_box_aspect(2)


plt.show()
```
